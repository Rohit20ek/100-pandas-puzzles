Classifying Emails as Spam or Not

You are tasked with building a spam detection system. Use features extracted from emails (e.g., word frequency) to classify them as spam (1) or not spam (0).

```python
import pandas as pd
data = pd.read_csv("SMSSpamCollection.csv", encoding='latin-1', sep='\t', header=None, names=['Label', 'Message'])
# data = data[['v1','v2']]
data.columns = ['Label', 'Message']

# encode target variable
data['Label']= data['Label'].map({'ham':0,'spam': 1})
# text vectorisation
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(stop_words='english', max_features=5000)
X = vectorizer.fit_transform(data['Message']).toarray()
y=data['Label']
```

Train- test split


```python
# Train - test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Rf training

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
rf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
